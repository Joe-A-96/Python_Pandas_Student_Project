#Loc and iloc

```python
#returns all rows but only name column
df.loc[0:5,"name"]
```

```python
#returns first 5 rows and takes name and class columns
df.loc[0:5,["name","class"]]
```

```python
#this returns all rows
df.loc[:, "name"]
```

```python
#returns all names up to index 10
df.loc[:10, "name"]
```

```python
#returns all rows from index 10
df.loc[10:, "name"]
```

```python
#returns all results for column index 1 and 2
df.iloc[:,[1,2]]
```

```python
#returns first 5 rows with the column index of 1 and 2
#in loc, stopping range is included in the output
df.iloc[0:5,[1,2]]
```

```python
# 5 rows and 2 columns
df.iloc[0:5, 0:2]
```

# How to filter Data

```python
#gets mark data that is greater than 90
df[df["mark"]>90]
```

```python
# boolean filtering as it shows all the results true or false if greater or less than 90
df["mark"]>90
```

```python
#returns all the students within class 6
df[df["class"]=="Six"]
```

## filter by multiple conditions

```python
# returns all the females with the mark 80 or above
df[(df["gender"]=="female") & (df["mark"]>=80)]
```

```python
# this returns all the females or all the students with a mark 85 or above
df[(df["gender"]=="female") | (df["mark"]>=85)]
```

```python
# this can be used to filter results, iloc can't be used to filter
df.loc[df["gender"]=="male", ["name","class"]]
```
